```python
def solution_wins(x, y):
  x_total = 0
  y_total = 0
  last_victor = "none"
  consecutive_wins = 0
  
  for i in range(10):

    if x[i] > y[i]:
      x_total += i + 1
      if last_victor == "x" and consecutive_wins == 2:
        x_total += sum(val for val in range(i+2,11))
        break
      else:
        consecutive_wins = consecutive_wins + 1 if last_victor == "x" else 1
        last_victor = "x"

    elif x[i] < y[i]:
      y_total += i + 1
      if last_victor == "y" and consecutive_wins == 2:
        y_total += sum(val for val in range(i+2,11))
        break
      else:
        consecutive_wins = consecutive_wins + 1 if last_victor == "y" else 1
        last_victor = "y"

    else:
      consecutive_wins = 0
      last_victor = "none"

  return x_total > y_total


#print(newstrat_wins([34, 33, 33, 0, 0, 0, 0, 0, 0, 0],[0, 34, 0, 1, 1, 1, 1, 1, 1, 60]))
#print("\n")
#print(newstrat_wins([10, 10, 10, 10, 20, 0, 0, 0, 20, 20],[20, 0, 10, 5, 10, 5, 10, 20, 0, 0]))
#print("\n")
#print(newstrat_wins([10, 10, 10, 10, 10, 10, 10, 10, 10, 10],[20, 5, 10, 10, 5, 10, 20, 10, 10, 10]))
        
```

```python
def win_rate(solution, allSolutions):
  wins = 0
  for i in range(allSolutions.shape[0]):
    wins = wins+1 if solution_wins(solution, allSolutions[i]) else wins 
  return wins/(allSolutions.shape[0])

def get_neighbor(solution):
  neighbor = solution.copy()
  nonzero_indices = []
  for i in range(10):
    if solution[i] >0:
      nonzero_indices.append(i)
  index_to_remove_from = random.choice(nonzero_indices)
  index_to_add_to = random.randint(0,9)
  neighbor[index_to_remove_from] = neighbor[index_to_remove_from] - 1
  neighbor[index_to_add_to] = neighbor[index_to_add_to] + 1
  return neighbor
  
def simulated_anneal(currentSolution, originalSolutions, improvedSolutions, overallBestRate, overallBestSolution ):
  modifiedSolution = get_neighbor(currentSolution)
  modifiedRate = (win_rate(modifiedSolution, originalSolutions) + win_rate(modifiedSolution, improvedSolutions))/(2)
  currentRate = (win_rate(currentSolution, originalSolutions) + win_rate(currentSolution, improvedSolutions))/(2)
  if modifiedRate > overallBestRate:
    overallBestRate = modifiedRate
    overallBestSolution = modifiedSolution
  if random.random() < .05 or modifiedRate > currentRate:
    currentSolution = modifiedSolution
    np.vstack([improvedSolutions , modifiedSolution])

  return currentSolution, originalSolutions, improvedSolutions, overallBestRate, overallBestSolution


```
